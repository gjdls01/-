```python
import gensim
ko_model = gensim.models.Word2Vec.load('/content/drive/MyDrive/ko.bin')
```

```python
from konlpy.tag import Kkma
from konlpy.utils import pprint
kkma = Kkma()
lst=[]
for i in range(len(ls)):
  a = kkma.pos(ls[i])
  lst.append(a)
lst
```

```python
for i in range(len(lst)):
  print(lst[i])
```

```python
w2v = []
for i in range(len(lst)):
  for j in lst[i]:
    a = []
    temp = ko_model.wv.most_similar(j[0])
    a.append([j[0],temp[0][0],temp[1][0],temp[2][0]])
    w2v.append(a)
```
